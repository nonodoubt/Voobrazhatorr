```python
import torch
import numpy as np
import cv2
import os
from PIL import Image
import matplotlib.pyplot as plt
from torchvision.transforms import functional as TF
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo

# Устройство
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def deeplabv3_segmentation(image, model_name, device="cuda"):
    if model_name == "resnet101":
        from torchvision.models.segmentation import deeplabv3_resnet101 as deeplab
    elif model_name == "resnet50":
        from torchvision.models.segmentation import deeplabv3_resnet50 as deeplab
    elif model_name == "mobilenet_v3":
        from torchvision.models.segmentation import deeplabv3_mobilenet_v3_large as deeplab
    else:
        raise ValueError("Unsupported DeepLab model.")

    model = deeplab(pretrained=True).to(device).eval()
    input_tensor = TF.to_tensor(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = model(input_tensor)['out']
    mask = output.argmax(1).squeeze(0).cpu().numpy()
    return mask

def detectron2_segmentation(image, model_name, device="cuda"):
    cfg = get_cfg()
    cfg.merge_from_file(model_zoo.get_config_file(model_name))
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
    cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(model_name)
    predictor = DefaultPredictor(cfg)

    outputs = predictor(np.array(image))
    masks = outputs["instances"].to("cpu").pred_masks.numpy()
    return masks.sum(axis=0)  # Объединяем все маски

# FastSAM - сегментация по точке
def fast_sam_segmentation(image, point_coords, device="cuda"):
    model = FastSAM("FastSAM-x.pt")
    
    everything_results = model(image, device=device, retina_masks=True, imgsz=1024, conf=0.4, iou=0.9,)
    prompt_process = FastSAMPrompt(image, everything_results, device=DEVICE)
    ann = prompt_process.everything_prompt()

    
    # Пример 1: Box prompt
    #box_annotations = prompt_process.box_prompt(bboxes=[[200, 200, 300, 300]])
    
    # Пример 2: Text prompt
    #text_annotations = prompt_process.text_prompt(text="a photo of a dog")
    
    # Пример 3: Point prompt
    #point_annotations = prompt_process.point_prompt(points=point_coords, pointlabel=[1])
    
    return ann

### УТИЛИТЫ

# Оценка затрат ресурсов
def evaluate_resources(func, *args, **kwargs):
    import time
    import psutil
    from torch.cuda import memory_allocated, memory_reserved

    start_time = time.time()
    start_mem = psutil.Process().memory_info().rss  # CPU memory
    start_cuda_mem = memory_allocated() if torch.cuda.is_available() else 0

    result = func(*args, **kwargs)

    end_time = time.time()
    end_mem = psutil.Process().memory_info().rss
    end_cuda_mem = memory_allocated() if torch.cuda.is_available() else 0
    print()
    print(str(func))
    print(f"Time: {end_time - start_time:.2f}s")
    print(f"CPU Memory Used: {(end_mem - start_mem) / 1024**2:.2f} MB")
    if torch.cuda.is_available():
        print(f"CUDA Memory Used: {(end_cuda_mem - start_cuda_mem) / 1024**2:.2f} MB")

    return result

# Экспорт маски
def export_mask(mask, file_path, invert=False):
    mask = 255 * mask.astype(np.uint8)
    if invert:
        mask = 255 - mask
    Image.fromarray(mask).save(file_path)

### ОСНОВНОЙ КОД

# Загрузка изображения
image_path = "/kaggle/input/testtest/PXL_20230917_123726955.jpg"
image = Image.open(image_path).convert("RGB")

# Выбор точек для FastSAM
#point_coords = [[1500, 1800]]  # Замените на координаты точки

# Применение моделей
masks = {
    "DeepLabv3-ResNet101": evaluate_resources(deeplabv3_segmentation, image, "resnet101", device),
    "DeepLabv3-ResNet50": evaluate_resources(deeplabv3_segmentation, image, "resnet50", device),
    "DeepLabv3-MobileNetV3": evaluate_resources(deeplabv3_segmentation, image, "mobilenet_v3", device),
    "Detectron2-MaskRCNN-R50": evaluate_resources(detectron2_segmentation, image, "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"),
    "Detectron2-MaskRCNN-R101": evaluate_resources(detectron2_segmentation, image, "COCO-InstanceSegmentation/mask_rcnn_R_101_FPN_3x.yaml"),
    #"FastSAM": evaluate_resources(fast_sam_segmentation, image, point_coords,device),
}

# Визуализация
plt.figure(figsize=(15, 10))
plt.subplot(2, 4, 1)
plt.imshow(image)
plt.title("Original Image")

for i, (name, mask) in enumerate(masks.items(), start=2):
    plt.subplot(2, 4, i)
    plt.imshow(mask, cmap="jet")
    plt.title(name)

plt.tight_layout()
plt.show()

# Экспорт масок
output_dir = "/kaggle/working/masks/"
os.makedirs(output_dir, exist_ok=True)

for name, mask in masks.items():
    export_mask(mask, os.path.join(output_dir, f"{name}.png"))
```
